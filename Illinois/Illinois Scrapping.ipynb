```python
from lxml import html
from lxml import etree
from os import listdir
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import pandas as pd
import numpy as np
import csv
```

```python
#Load IL zip codes
ILZip = list(pd.read_csv('ILZip.csv')['ZIP'])
```

```python
current_path= !cd
chrome_path = r"C:\Users\amartinezcotto\Desktop\chromedriver_win32\chromedriver.exe"
chromeOptions = webdriver.ChromeOptions()
prefs = {"download.default_directory" : current_path[0]}
chromeOptions.add_experimental_option("prefs",prefs)
driver = webdriver.Chrome(executable_path=chrome_path, chrome_options=chromeOptions)
```

```python
## Loop in the same window
license_data = []
for ZipCode in ILZip:
    driver.get('https://www.illinois.gov/ilcc/SitePages/LicenseLookup.aspx?Zip='+str(ZipCode)+'&Status=A&')
    while True:
        try:
            wait = WebDriverWait(driver, 10)
            element = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="{E55DE6F9-991D-4CA5-91AC-B64892CBBC45}-{2CB7CF0E-952E-48D0-A50F-B506971F1FD4}"]/tbody/tr[2]/td[2]')))
        except:
            break
        licenses = driver.find_elements_by_class_name('ms-vb2')
        for data in licenses:
            license_data.append(data.text)
        try:
            driver.find_element_by_xpath('''//*[@id="bottomPagingCellWPQ4"]/table/tbody/tr/td[2]/a/img''').click()
        except:
            break
                
```
